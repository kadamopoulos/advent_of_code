```python
def get_digit(arr, pos):
    trunc = arr[:-pos]
    if pos == 0:
        trunc = arr
        
    amax = np.argmax(trunc)
    return(trunc[amax], arr[(amax+1):])
```

```python
data = [[int(i) for i in line] for line in input_data]

joltages = []

for bank in data:
    joltages.append([])
    for i in range(11, -1, -1):
        digit, bank = get_digit(bank, i)
        joltages[-1].append(int(digit))

sum([int("".join([str(digit) for digit in joltage])) for joltage in joltages])
```

```python
def get_joltage(data):
    data_2 = [[int(i) for i in line] for line in data]

    tens_array = np.array(data_2)[:, :-1]

    tens = np.max(tens_array, axis=1)
    pos = np.argmax(tens_array, axis=1)

    ones_array = [np.array(data_2[i][(pos[i]+1):]) for i in range(len(data_2))]
    
    ones = [max(array) for array in ones_array]

    return int(sum(tens)*10 + sum(ones))

get_joltage(input_data)
```
