```python
with open('input_1.txt', 'r') as file:
    input_data = file.read().splitlines()
```

```python
def origin_points(initial_location):
    origin_counter = 0
    current_location = initial_location
    for rotation in input_data:
        if rotation[0] == 'R':
            current_location += int(rotation[1:])
        else:
            current_location -= int(rotation[1:])
        
        if current_location % 100 == 0:
            origin_counter += 1
    
    return origin_counter
```
