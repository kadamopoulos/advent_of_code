```python
with open('input_1.txt', 'r') as file:
    input_data = file.read().splitlines()
```

```python
with open('test.txt', 'r') as file:
    test_data = file.read().splitlines()
```

```python
def rotate(location, rotation):
    if rotation[0] == 'R':
        location += int(rotation[1:])
    else:
        location -= int(rotation[1:])
    return location
```

```python
def origin_points(initial_location, data = input_data):
    origin_counter = 0
    current_location = initial_location
    for rotation in data:
        current_location = rotate(current_location, rotation)
        
        if current_location % 100 == 0:
            origin_counter += 1
    
    return origin_counter
```

```python
origin_points(50, test_data)
```

```python
origin_points(50)
```

```python
def origin_passes(initial_location, data = input_data):
    origin_counter = 0
    current_location = initial_location
    was_0 = False
    for rotation in data:
        sign = rotation[0]
        distance = int(rotation[1:])
        
        origin_counter += distance//100

        current_location = rotate(current_location, sign+str(distance%100))
        current_location = current_location%200
        if current_location >= 100:
            origin_counter += 1
            current_location -= 100

        if was_0:
            if sign == "L":
                origin_counter -= 1
            was_0 = False

        if current_location == 0:
            if sign == "L":
                origin_counter += 1
            was_0 = True
                  
        print(current_location, origin_counter)

    return origin_counter
```

```python
origin_passes(50, test_data)
```

```python
origin_passes(50)
```
